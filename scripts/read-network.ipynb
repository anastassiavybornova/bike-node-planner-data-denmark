# Read in simplified knudepunkter network

Sample code to read in gdf of simplified network (as generated by [`generate-network-communication.ipynb`](./generate-network-communication.ipynb)), and convert it into a network object

```python
import geopandas as gpd
import momepy
import networkx as nx
import folium
```

```python
# read in gdfs
gdfs = gpd.read_file("../data/network-communication/gdfs.gpkg")


# convert with momepy
G = momepy.gdf_to_nx(
    gdf_network = gdfs, 
    multigraph=False, 
    integer_labels=True
)
```

```python
# count connected components
print(len([c for c in nx.connected_components(G)]))
```

```python
# using momepy to get nodes and edges gdf with corresponding labels and geometry objects
nodes, edges = momepy.nx_to_gdf(
    net = G,
    points = True,
    lines = True
)
```

```python
m=edges.explore(tiles = "cartodb positron", prefer_canvas=True, name = "edges", style_kwds={"weight":1}, column = "comp", cmap = "Dark2")
nodes.explore(m=m, name = "nodes", marker_kwds={"radius":3}, color = "orange")
folium.LayerControl().add_to(m)
m
```
